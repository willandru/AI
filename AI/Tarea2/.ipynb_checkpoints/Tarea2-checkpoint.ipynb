### BERNOULLI con cierto p entre 0 y 1

```python
# DISTRIBUCIÓN BERNOULLI
p=0.3 

# Los primeros 4 momentos estadisticos de esta muestra se pueden calcular utilizando 'mvsk'

# Calculamos los valores TEÓRICOS
mean, var, skew, kurt = bnl.stats(p, moments='mvsk') 

#Se realiza un muestreo aleatorio de la distribucion utilizando 'rvs'
r= bnl.rvs(p, size=N)

#Calculamos los valores EXPERIMENTALES
avgest=r.mean()
varest=r.var()

print("Los valores TEORICOS son: varianza="+ str(var) + " , media="+str(mean))
print("Los valores EXPERIMENTALES son: varianza="+ str(varest) + " , media="+str(avgest))
```

```python
# Creamos un subplot
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
bins=int(np.sqrt(N))


ax[0].plot(r)
ax[0].set_title(str(N) + " Puntos muestreados")
r.sort()
ax[1].plot(r)
ax[1].set_title('Puntos Ordenados')
ax[2].hist(r, bins )
ax[2].set_title('HISTOGRAMA' + str(bins))


fig.suptitle('Distribución Bernoulli p='+ str(p))
```

# Punto 2. Usando las columnas de "price" como etiqueta y "carat" como características del archivo 'diamonds.csv', realice una regresión lineal y resuelva el par de parametros de la hipótesis mediante el método de fuerza bruta "grid search", puede usar como base el código uministrado: 
###  https://colab.research.google.com/drive/1weuKWSSNK0oyxg3jgv1xYzz9sORU7pr4?usp=sharing

## 1. Plantee un método para obtener una solución aproximada (fuerza bruta), explíquelo en no más de 4 parrafos.
## 2. Solucione el problema
## 3. De sus conclusiones


.

```python
import pandas as pd

data = pd.read_csv('diamonds.csv')

X=data["carat"].values
Y=data["price"].values
```

```python
# Define a range of values for intercept and coefficient to search over
intercepts = np.linspace(-7000, -1000, 1000)
coefficients = np.linspace(3000, 10000, 1000)


best_intercept = None
best_coef = None
min_mse = float('inf')

# Loop over all possible combinations of intercept and coefficient values
for intercept in intercepts:
    for coef in coefficients:
        
        # Calculate the predicted values of y for the current intercept and coefficient
        y_pred = intercept + coef*X
        
        # Calculate the mean squared error between the predicted and actual values of y
        mse = np.mean((y_pred - Y)**2)
        
        # Update the best intercept, coefficient and minimum MSE if the current MSE is lower
        if mse < min_mse:
            best_intercept = intercept
            best_coef = coef
            min_mse = mse

# Print the best intercept, coefficient and minimum MSE
print('Best intercept:', best_intercept)
print('Best coefficient:', best_coef)
print('Minimum MSE:', min_mse)
```

```python
y_predict= best_intercept + X*best_coef

plt.scatter(X, Y, color='blue')
plt.plot(X, y_predict, color='red', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scattered data and fitting line')
plt.show()
```

```python
def costo(theta0,theta1,x,y):
  #donde theta0 y theta1 son los parámetros a probar,
  #x y y son nparrays de la misma dimensión
  n=x.size
  #aqui podrian depurar que los tamaños de los arreglos sean iguales

  #calculo del costo en la variable J
  J=(1/(2*n))*np.sum(np.power(((theta0+theta1*x))-y,2)) #terminen ustedes... 
  #time.sleep(1)#ahora lo activan para que se vea bonito y pause la ejecucion
  return J

def gradientetheta0(theta0,theta1,x,y):
  #donde theta0 y theta1 son los parámetros a probar,
  #x y y son nparrays de la misma dimensión
  #ver https://numpy.org/doc/stable/reference/generated/numpy.array.html
  n=x.size
  #aqui podrian depurar que los tamaños de los arreglos sean iguales
  return (1/n)*np.sum((theta0+theta1*x)-y)#terminen acà el gradiente 0

def gradientetheta1(theta0,theta1,x,y):
  #donde theta0 y theta1 son los parámetros a probar,
  #x y y son nparrays de la misma dimensión
  #ver https://numpy.org/doc/stable/reference/generated/numpy.array.html
  n=x.size
  #aqui podrian depurar que los tamaños de los arreglos sean iguales
  return (1/n)*np.sum(((theta0+theta1*x)-y)*x)#terminen acà el gradiente 1 

def PasoDescensoGradiente(alpha,theta0,theta1,x,y):
  #Donde alpha es el coeficiente que multiplica al gradiente
  #donde theta0 y theta1 son los parámetros a probar,
  #la salida es una pareja thetaK0 y thetaK1 son los parámetros del siguiente salto
  N1=gradientetheta0(theta0,theta1,x,y)#calculen gradiente 0
  N2=gradientetheta1(theta0,theta1,x,y)#Calculen gradiente 1
  thetaK0=theta0-alpha*N1#haga el siguiente paso igual a el punto donde estoy, menos alpha veces el gradiente calculado en este caso para thera 0 y theta 1
  thetaK1=theta1-alpha*N2# ojo que no es thetaK1=theta1-alpha*gradientetheta1(thetaK0,theta1,x,y)
  
  return thetaK0,thetaK1
```
